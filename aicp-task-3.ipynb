```python
# Source data (two-dimensional array)
electricity_matrix = [
    [50, 150, 250],
    [100, 200, 300]
]

# Function to calculate and display cost for slab 1
def costSlab1():
    unit_range = 100
    cost_per_unit = 10
    units_consumed = electricity_matrix[0][0]
    total_cost = units_consumed * cost_per_unit
    print(f"Slab 1 - Units: {units_consumed}, Cost: Rs.{total_cost}")

# Function to calculate and display cost for slab 2
def costSlab2():
    unit_range = 100
    cost_per_unit = 15
    units_consumed = electricity_matrix[1][0] - electricity_matrix[0][0]
    total_cost = units_consumed * cost_per_unit
    print(f"Slab 2 - Units: {units_consumed}, Cost: Rs.{total_cost}")

# Function to calculate and display cost for slab 3
def costSlab3():
    unit_range = 100
    cost_per_unit = 20
    units_consumed = electricity_matrix[1][1] - electricity_matrix[0][1]
    total_cost = units_consumed * cost_per_unit
    print(f"Slab 3 - Units: {units_consumed}, Cost: Rs.{total_cost}")

# Main menu
while True:
    student_id = input("Enter your student ID: ")
    print(f"Student ID: {student_id}")

    print("Menu:")
    print("1. Display the bill of slab 1 and slab 2")
    print("2. Display the bill of slab 3")
    print("Any other key to exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        costSlab1()
        costSlab2()
    elif choice == '2':
        costSlab3()
    else:
        print("Exiting the program.")
        break
```
